# Collections & Comprehensions



## Part 1 ‚Äî Sequences

## The sequence protocol

A Python *sequence* is any object that supports:

- `len(x)` ‚Äî how many elements
- `x[i]` ‚Äî indexed access (0-based; negative indices count from the end)
- `x[i:j]` and `x[i:j:k]` ‚Äî slicing
- `for item in x` ‚Äî iteration
- `item in x` ‚Äî membership

`list`, `tuple`, `range`, `str`, `bytes`, `bytearray` all satisfy it. That's why the same `for x in xs`, `xs[0]`, `xs[1:5]` syntax works across all of them.

Sequences split into **mutable** (`list`, `bytearray`) and **immutable** (`tuple`, `str`, `bytes`, `range`). Same protocol, different behavior for in-place operations.

## `list` ‚Äî the mutable workhorse

`list` is the everyday container. Heterogeneous (any types), mutable, ordered.

Construction:

```python
[]                              # empty
[1, 2, 3]                       # literal
list("abc")                     # ['a', 'b', 'c']  ‚Äî from any iterable
list(range(5))                  # [0, 1, 2, 3, 4]
```

The methods you'll use daily:

| Method | What it does | Cost |
|---|---|---|
| `xs.append(x)` | add one item to the end | O(1) amortized |
| `xs.extend(iter)` | add many items from any iterable | O(k) |
| `xs.insert(i, x)` | insert at index `i` | O(n) ‚Äî shifts the tail |
| `xs.pop()` / `pop(i)` | remove + return last (or i-th) | O(1) / O(n) |
| `xs.remove(x)` | remove first occurrence of `x` | O(n) |
| `xs.index(x)` | first index of `x`, or raise | O(n) |
| `xs.count(x)` | how many times `x` appears | O(n) |
| `xs.sort()` / `xs.reverse()` | in place | O(n log n) / O(n) |
| `sorted(xs)` / `reversed(xs)` | new list / iterator | non-mutating |
| `xs.copy()` / `xs[:]` | shallow copy | O(n) |

````python
xs = [1, 2, 3]

xs.append(4)
print(xs)              # [1, 2, 3, 4]

xs.extend([5, 6])
print(xs)              # [1, 2, 3, 4, 5, 6]

xs.insert(0, 99)
print(xs)              # [99, 1, 2, 3, 4, 5, 6]

last = xs.pop()
print(last, xs)        # 6 [99, 1, 2, 3, 4, 5]

xs.remove(99)
print(xs)              # [1, 2, 3, 4, 5]

# sort vs sorted ‚Äî in place vs returning a new list
xs = [3, 1, 4, 1, 5, 9, 2, 6]
xs.sort()
print(xs)              # mutated in place

ys = [3, 1, 4, 1, 5, 9, 2, 6]
sy = sorted(ys)
print(ys)              # unchanged
print(sy)              # new sorted list

# Sort with key= ‚Äî by derived field
words = ["alpha", "bee", "delta"]
print(sorted(words, key=len))             # by length
print(sorted(words, key=str.lower, reverse=True))
````

## Indexing and slicing

The full slice form is `xs[start:stop:step]`. All three are optional. Negative indices count from the end. The shape:

| Slice | Meaning |
|---|---|
| `xs[i]` | the item at index `i` (negative ok) |
| `xs[i:j]` | items from `i` (inclusive) to `j` (exclusive) |
| `xs[i:j:k]` | every `k`-th item in that range |
| `xs[:n]` | the first `n` items |
| `xs[n:]` | from `n` onward |
| `xs[-n:]` | the last `n` items |
| `xs[::-1]` | the whole sequence reversed |
| `xs[::2]` | every other item |

Slicing **always returns a new object of the same type** (a new `list` from a list, a new `str` from a string, etc.). Out-of-range indices raise `IndexError`; out-of-range *slices* are silently clipped ‚Äî `xs[100:200]` on a short list is `[]`, not an error.

````python
xs = [10, 20, 30, 40, 50, 60, 70, 80]

print(xs[0])           # 10
print(xs[-1])          # 80 ‚Äî last
print(xs[-3])          # 60 ‚Äî third from end

print(xs[2:5])         # [30, 40, 50]
print(xs[:3])          # [10, 20, 30]
print(xs[5:])          # [60, 70, 80]
print(xs[-3:])         # [60, 70, 80]
print(xs[::2])         # [10, 30, 50, 70]
print(xs[::-1])        # reversed

# Out-of-range slice ‚Äî silently clipped
print(xs[100:200])     # []

# Slice assignment ‚Äî mutates a sub-range
xs[2:5] = ["a", "b", "c", "d"]
print(xs)              # [10, 20, 'a', 'b', 'c', 'd', 60, 70, 80] ‚Äî length can change

# Delete via slice
del xs[1:4]
print(xs)              # [10, 'b', 'c', 'd', 60, 70, 80]
````

## Unpacking ‚Äî the `*` in assignment and literals

Python's unpacking syntax composes nicely once you've seen it once.

```python
a, b, c = [1, 2, 3]              # unpack a 3-element iterable
a, *rest = [1, 2, 3, 4, 5]       # a=1, rest=[2, 3, 4, 5]
first, *middle, last = [1, 2, 3, 4, 5]  # first=1, middle=[2, 3], last=5

# Inside literals ‚Äî splat into a new collection
combined = [*xs, *ys, 99]        # concatenate xs and ys into a new list
merged   = {*set_a, *set_b}      # union into a new set
all_pairs = {**dict_a, **dict_b} # merge dicts (right wins on duplicates)
```

The `*name` form on the **left** side of `=` is sometimes called "starred assignment." On the **right** side of `=`, or inside a literal, it is "splat" or "unpacking." Same star, different meaning by position.

````python
# Starred assignment
first, *middle, last = [1, 2, 3, 4, 5]
print(first, middle, last)     # 1 [2, 3, 4] 5

# Swap ‚Äî works because the right side is an unpacked tuple
a, b = 1, 2
a, b = b, a
print(a, b)                    # 2 1

# Splat inside literals
xs = [1, 2]
ys = [3, 4]
combined = [*xs, *ys, 99]
print(combined)                # [1, 2, 3, 4, 99]

merged = {*{1, 2, 3}, *{3, 4, 5}}
print(merged)                  # {1, 2, 3, 4, 5}

# Dict merge ‚Äî 3.5+ with **; 3.9+ has the | operator
a = {"x": 1, "y": 2}
b = {"y": 99, "z": 3}
print({**a, **b})              # {'x': 1, 'y': 99, 'z': 3}
print(a | b)                   # same (3.9+)
````

## `tuple` ‚Äî the immutable cousin

`tuple` is the immutable sequence type. Same indexing and slicing as `list`; no methods that mutate.

Two distinct uses, often conflated:

- **Record-like** ‚Äî a fixed-shape group of values, each position meaning something. `(host, port, ssl)`. When you reach for tuple here and find yourself naming positions in comments, switch to a `dataclass` or `NamedTuple` (notebook 05) for clarity.
- **Frozen sequence** ‚Äî a list-like collection you don't want anyone to mutate. Useful as a dict key (lists can't be keys; tuples can, if their elements are hashable).

Literal form: parentheses, or just commas (the comma is what makes a tuple, not the parens). A single-element tuple needs a trailing comma: `(42,)`. `(42)` is just the integer 42 in parentheses.

````python
# Construction
empty = ()
single = (42,)                 # the comma is required
multi = (1, 2, 3)
without_parens = 1, 2, 3       # also a tuple ‚Äî the commas are what matter
print(type(without_parens))    # <class 'tuple'>

# Indexing and slicing ‚Äî same as list
print(multi[0])
print(multi[1:])

# Immutability ‚Äî no append, no item assignment
# multi[0] = 99    # TypeError

# Tuples can be dict keys; lists cannot
d = {(1, 2): "ne", (-1, -1): "sw"}
print(d[(1, 2)])

# Multiple return values are tuples
def divmod_simple(a, b):
    return a // b, a % b       # commas make a tuple

q, r = divmod_simple(17, 5)
print(q, r)
````

## `str` ‚Äî sequence of characters

`str` is also a sequence, but with so many string-specific methods that it deserves its own focus. Notebook 02 covered literals and f-strings; here are the methods you'll reach for daily.

| Method | What it does |
|---|---|
| `s.split(sep)` | split on `sep` (default: any whitespace) |
| `sep.join(iter)` | join with `sep` as the separator |
| `s.strip()` / `lstrip` / `rstrip` | remove leading/trailing whitespace |
| `s.replace(old, new)` | substring replace |
| `s.startswith(p)` / `endswith(p)` | prefix/suffix test |
| `s.upper()` / `lower()` / `title()` | case conversion |
| `s.find(p)` / `index(p)` | first index of `p`, or -1 / raise |
| `s.count(p)` | how many non-overlapping `p` |
| `s.format(...)` | older interpolation ‚Äî prefer f-strings now |
| `s.splitlines()` | split on line endings |

Strings are immutable ‚Äî these all return new strings.

````python
s = "the quick brown fox"

print(s.split())                       # ['the', 'quick', 'brown', 'fox']
print(s.split(maxsplit=1))             # ['the', 'quick brown fox']
print(",".join(["a", "b", "c"]))       # 'a,b,c'

print("  hello  ".strip())             # 'hello'
print("hello, world".replace("hello", "hi"))

print(s.startswith("the"))             # True
print(s.endswith("fox"))               # True

print(s.upper())                       # 'THE QUICK BROWN FOX'
print(s.title())                       # 'The Quick Brown Fox'

print(s.find("brown"))                 # 10
print(s.find("zebra"))                 # -1

# split + join ‚Äî a common transformation idiom
csv = "alpha, beta, gamma"
parts = [p.strip() for p in csv.split(",")]
print(parts)                           # ['alpha', 'beta', 'gamma']
print("|".join(parts))                 # 'alpha|beta|gamma'
````

## Part 2 ‚Äî Keyed collections

## `dict` ‚Äî the mapping workhorse

`dict` maps keys to values. Keys must be **hashable** (immutable types: numbers, strings, tuples-of-hashables, frozensets). Values can be anything.

Since Python 3.7, dicts **preserve insertion order**. This was a CPython implementation detail in 3.6 and became a language guarantee in 3.7.

Lookup has two forms ‚Äî unsafe (`d[key]`, raises `KeyError`) and safe (`d.get(key)`, returns `None` or a default):

| Form | If key missing |
|---|---|
| `d[key]` | raises `KeyError` |
| `d.get(key)` | returns `None` |
| `d.get(key, default)` | returns `default` |
| `d.setdefault(key, default)` | inserts and returns `default` if missing |

````python
ages = {"alice": 30, "bob": 25, "carol": 42}

# Unsafe ‚Äî raises on missing
print(ages["alice"])
# print(ages["dave"])    # KeyError

# Safe ‚Äî returns None or default
print(ages.get("alice"))         # 30
print(ages.get("dave"))          # None
print(ages.get("dave", -1))      # -1

# Iteration ‚Äî keys by default
for name in ages:
    print(name)

# Iteration ‚Äî values
for age in ages.values():
    print(age)

# Iteration ‚Äî items (key, value pairs)
for name, age in ages.items():
    print(f"{name} is {age}")

# Membership tests the keys
print("alice" in ages)           # True
print(30 in ages)                # False ‚Äî 30 is a value, not a key
print(30 in ages.values())       # True
````

## Updating dicts ‚Äî `update`, `setdefault`, the `|` merge

`dict.update(other)` merges another dict into this one (in place). `dict.setdefault(k, v)` is the "if not present, set; either way, return the value" idiom.

Python 3.9+ adds `|` (merge into a new dict) and `|=` (merge in place). Right-hand wins on duplicates.

````python
a = {"x": 1, "y": 2}
b = {"y": 99, "z": 3}

# update ‚Äî in place
a.update(b)
print(a)                  # {'x': 1, 'y': 99, 'z': 3}

# | ‚Äî new dict
a = {"x": 1, "y": 2}
print(a | b)              # {'x': 1, 'y': 99, 'z': 3} ‚Äî a unchanged
print(a)                  # {'x': 1, 'y': 2}

# |= ‚Äî in place (same as update)
a |= b
print(a)

# setdefault ‚Äî useful for grouping
groups = {}
for word in ["apple", "ant", "bee", "berry", "bat"]:
    groups.setdefault(word[0], []).append(word)
print(groups)             # {'a': ['apple', 'ant'], 'b': ['bee', 'berry', 'bat']}
````

## `collections` ‚Äî the specialized containers

The `collections` module ships specialized dict and tuple variants for common patterns.

- **`defaultdict(factory)`** ‚Äî a dict where missing keys auto-construct via the factory. Avoids `setdefault` boilerplate.
- **`Counter(iter)`** ‚Äî count occurrences of each element. Iterable in, dict of counts out.
- **`OrderedDict`** ‚Äî preserves insertion order. Mostly obsolete since regular dicts do that since 3.7; still has a couple unique methods (`move_to_end`).
- **`namedtuple("Name", [...])`** ‚Äî a tuple with named fields. Lightweight when a full `dataclass` would be overkill.

````python
from collections import defaultdict, Counter, namedtuple

# defaultdict ‚Äî group without setdefault
groups = defaultdict(list)
for word in ["apple", "ant", "bee", "berry", "bat"]:
    groups[word[0]].append(word)
print(dict(groups))

# Counter ‚Äî count occurrences
text = "the rain in spain stays mainly in the plain"
counts = Counter(text.split())
print(counts)                  # Counter({'the': 2, 'in': 2, ...})
print(counts.most_common(3))   # top three

# namedtuple ‚Äî lightweight record
Point = namedtuple("Point", ["x", "y"])
p = Point(3, 4)
print(p)                       # Point(x=3, y=4)
print(p.x, p.y)                # 3 4
print(p[0])                    # 3 ‚Äî also indexable like a regular tuple
````

## `set` and `frozenset`

A `set` is an unordered collection of unique, hashable elements. Membership is O(1).

The set operators map directly to set algebra:

| Operator | Method | Meaning |
|---|---|---|
| `\|` | `union` | in either |
| `&` | `intersection` | in both |
| `-` | `difference` | in left, not in right |
| `^` | `symmetric_difference` | in one or the other, not both |
| `<=` / `<` | `issubset` / proper subset | every element of left in right |
| `>=` / `>` | `issuperset` / proper superset | the reverse |

`frozenset` is the immutable variant ‚Äî same operations, no mutation. Required when you need a set as a dict key or as an element of another set.

````python
warm = {"red", "orange", "yellow"}
cool = {"blue", "green", "violet"}
primary = {"red", "blue", "yellow"}

print(warm | cool)          # union
print(warm & primary)       # intersection: red, yellow
print(warm - primary)       # difference: orange
print(warm ^ primary)       # symmetric diff: orange + blue

print(warm <= warm | cool)  # True
print({"red"} <= warm)      # True

# Set methods that mutate
s = {1, 2, 3}
s.add(4)
s.discard(99)               # no error if absent
s.remove(2)                 # KeyError if absent
print(s)

# frozenset ‚Äî hashable, can be set element / dict key
fs = frozenset([1, 2, 3])
container = {fs, frozenset([4, 5])}
print(container)
````

## Hashability ‚Äî what makes a key valid

A key in a `dict`, or an element in a `set`, must be **hashable**. An object is hashable if:

- It has a `__hash__` method (most built-in immutable types do).
- Equal objects produce equal hashes ‚Äî `a == b` must imply `hash(a) == hash(b)`.

Practical version: **immutable built-in types are hashable**. Mutable types (`list`, `dict`, `set`) are not ‚Äî they'd be a hazard, because mutating after insertion would corrupt the hash bucket.

Custom classes (notebook 05) are hashable by default (using `id()`), but if you override `__eq__` you should also override `__hash__` to keep the contract.

````python
# Hashable ‚Äî can be a set element or dict key
print(hash(42))
print(hash("hello"))
print(hash((1, 2, 3)))            # tuples-of-hashables work

# Not hashable
try:
    hash([1, 2, 3])
except TypeError as e:
    print(f"list:  {e}")

try:
    hash({1, 2})
except TypeError as e:
    print(f"set:   {e}")

# But a frozenset IS hashable
print(hash(frozenset([1, 2, 3])))
````

## Part 3 ‚Äî Comprehensions

## List comprehensions ‚Äî the shape

A **list comprehension** builds a new list by transforming/filtering an existing iterable, in one expression.

```python
[expression  for x in iterable  if condition]
```

Reads as: *"for each `x` in `iterable`, if `condition`, evaluate `expression` and collect into a new list."*

This pattern replaces almost every short `for`-and-append loop. Use a comprehension when the body is one expression. Reach for an explicit loop when there's branching, side effects, or multiple statements per element.

````python
# Square each number
squares = [x * x for x in range(10)]
print(squares)

# With a filter ‚Äî only even squares
even_squares = [x * x for x in range(10) if x % 2 == 0]
print(even_squares)

# Transform a list of names
names = ["alice", "bob", "carol"]
upper = [n.upper() for n in names]
print(upper)

# Compare against the explicit-loop equivalent
result = []
for n in names:
    result.append(n.upper())
print(result)
# Comprehension is the same thing; one expression instead of three lines
````

## Set and dict comprehensions

Same shape, different surrounding brackets ‚Äî `{}` (no key:value) gives a **set comprehension**; `{k: v ...}` gives a **dict comprehension**.

````python
# Set comprehension ‚Äî unique squares (note duplicates removed)
nums = [1, -1, 2, -2, 3, -3]
unique_squares = {n * n for n in nums}
print(unique_squares)        # {1, 4, 9}

# Dict comprehension ‚Äî build a dict from an iterable
names = ["alice", "bob", "carol"]
name_lengths = {n: len(n) for n in names}
print(name_lengths)

# With a filter
ages = {"alice": 30, "bob": 25, "carol": 42, "dave": 17}
adults = {name: age for name, age in ages.items() if age >= 18}
print(adults)

# Invert a dict ‚Äî swap keys and values
flipped = {v: k for k, v in ages.items()}
print(flipped)
````

## Generator expressions ‚Äî the parenthesized cousin

A **generator expression** has the same shape as a list comprehension but with `()` instead of `[]`. The difference is huge: a generator expression is **lazy** ‚Äî it produces values one at a time as you iterate, without materializing the whole sequence in memory.

```python
[x * x for x in range(10**8)]   # builds a 100 million-element list ‚Äî OOM risk
(x * x for x in range(10**8))   # builds a generator ‚Äî uses constant memory
```

Use a generator expression when you'll iterate the result once and don't need indexed access. Many functions (`sum`, `min`, `max`, `any`, `all`, `join`) accept any iterable, so a generator works directly:

```python
sum(x * x for x in range(10**6))     # parentheses can be omitted as a sole argument
```

If you need a list (because you'll index, slice, or iterate twice), use a list comprehension. If you'll iterate once, generators are usually the better default.

````python
# Generator expression ‚Äî lazy
gen = (x * x for x in range(5))
print(gen)                              # <generator object ...>
print(list(gen))                        # materialize once: [0, 1, 4, 9, 16]
print(list(gen))                        # empty! generators are single-pass

# As a function argument ‚Äî parentheses can be omitted
total = sum(x * x for x in range(10))
print(total)                            # 285

# Any/all ‚Äî short-circuit, generator-friendly
print(any(x > 100 for x in [1, 2, 200, 3]))   # True
print(all(x > 0 for x in [1, 2, 3]))          # True

# Joining strings ‚Äî common use
print(", ".join(str(x) for x in [1, 2, 3, 4]))
````

## Nested comprehensions

Comprehensions can have multiple `for` clauses (and `if` guards anywhere). The leftmost `for` is the outer loop; the rightmost is the inner. The order is "as you'd write it as nested loops."

```python
# All (x, y) pairs from two lists
pairs = [(x, y) for x in [1, 2, 3] for y in ["a", "b"]]
# Equivalent to:
# pairs = []
# for x in [1, 2, 3]:
#     for y in ["a", "b"]:
#         pairs.append((x, y))
```

Up to two levels of nesting is usually still readable. Past that, switch to explicit loops ‚Äî the comprehension becomes write-only.

````python
# Cartesian product
pairs = [(x, y) for x in [1, 2, 3] for y in ["a", "b"]]
print(pairs)

# Flatten a list of lists
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat = [n for row in matrix for n in row]
print(flat)

# Transpose a matrix ‚Äî comprehension over zip
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed = [list(col) for col in zip(*matrix)]
print(transposed)

# Filter on the inner loop
even_pairs = [(x, y) for x in range(4) for y in range(4) if (x + y) % 2 == 0]
print(even_pairs)
````

## Walrus in a comprehension

The walrus operator (`:=`, notebook 02) shines inside comprehensions when you want to compute a value once and reuse it in both the filter and the projection:

````python
import math

# Without walrus ‚Äî sqrt computed twice per element
result1 = [math.sqrt(x) for x in range(10) if math.sqrt(x) > 1.5]
print(result1)

# With walrus ‚Äî sqrt computed once
result2 = [y for x in range(10) if (y := math.sqrt(x)) > 1.5]
print(result2)
````

## When NOT to comprehensionify

Comprehensions are clean *only* when they're short and the intent is obvious. The cliff:

- **Side effects.** Don't write `[print(x) for x in xs]` to print things; that's a loop. The comprehension allocates a list of `None`s for no reason.
- **Multi-line bodies.** If you find yourself spanning three or four lines or chaining multiple `if` guards plus an `else`, fall back to a `for` loop. It will be more readable.
- **Multiple steps.** A comprehension that does three transformations in sequence often hides bugs. Break it into named intermediates.
- **`if/else` for the *projection*** uses the conditional expression form, not the `if` filter: `[x if x > 0 else 0 for x in xs]` ‚Äî but two of those nested is hard to read.

The honest rule: if you can read the comprehension out loud as a sentence and it makes sense, keep it. Otherwise, use a loop.
